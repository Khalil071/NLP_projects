```python
# Code for Topic Classification by using naive_bayes, Word2Vec, FastText

# by using naive_bayes
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score

# Splitting Data
X_train, X_test, y_train, y_test = train_test_split(df['Article_content'], df['Topic'], test_size=0.2, random_state=42)

# Creating a Pipeline
naive_bayes_classifier = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=500)),  # Convert text to numerical representation
    ('clf', MultinomialNB())  # Train Naïve Bayes Classifier
])

# Train the model
naive_bayes_classifier.fit(X_train, y_train)

# Predict & Evaluate
y_pred_NB = naive_bayes_classifier.predict(X_test)
Naive_Bayes_accuracy = accuracy_score(y_test, y_pred_NB)
print("Naive_Bayes Model Accuracy: {:.4f}".format(Naive_Bayes_accuracy))
print("Topic Classification Report:\n", classification_report(y_test, y_pred_NB))
```

```python
# by using Word2Vec, FastText
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from gensim.models import FastText, Word2Vec
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

nltk.download('stopwords')
nltk.download('punkt')
```

```python
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    tokens = word_tokenize(text.lower())  # Tokenization and lowercase
    tokens = [word for word in tokens if word not in stopwords.words('english')]  # Remove stopwords
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(word) for word in tokens]  # Stemming
    return tokens

# Apply preprocessing
df['processed_content'] = df['Article_content'].apply(preprocess_text)

# Train FastText and Word2Vec models
fasttext_model = FastText(sentences=df['processed_content'], vector_size=100, window=5,
                          min_count=1, workers=4)

word2vec_model = Word2Vec(sentences=df['processed_content'], vector_size=100, window=5,
                           min_count=1, workers=4)
```

```python
# Convert text data to numerical representation

def vectorize_text(text, model):
    vectors = [model.wv[word] for word in text if word in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)

df['FastText_Vectorized_Content'] = df['processed_content'].apply(lambda x: vectorize_text(x, fasttext_model))
df['Word2Vec_Vectorized_Content'] = df['processed_content'].apply(lambda x: vectorize_text(x, word2vec_model))
```
